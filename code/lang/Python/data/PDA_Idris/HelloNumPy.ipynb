## Basics

## Stacking

```python
a = np.arange(9).reshape(3,3)
print a 
b = 2*a
print b
print '\n'
```

### Horizontal stacking

```python
# Horizontal stacking
print '#1 hstack'
print np.hstack((a,b))
print np.concatenate((b,a), axis=1)
print '\n'
```

### Vertical stacking

```python
# Vertical stacking
print '#2 vstack'
print np.vstack((a,b))
print np.concatenate((b,a), axis=0)
print '\n'
```

```python
### Depth stacking
```

```python
# Depth stacking
print '#3 dstack'
print np.dstack((a,b))
print '\n'
```

### Column/Row stacking

```python
# Column stacking
print '#4 column_stack()'
c = np.array([0, 1])
d = np.array([2, 3])
print np.column_stack((c,d)) 
# 2D arrays are stacked the way the hstack() stacks them.
print np.column_stack((a,b)) == np.hstack((a,b))
print '\n'

# Row stacking
print '#5 row_stack()'
print np.row_stack((c,d))
# 2D arrays are stacked the way the vstack() stacks them.
print np.row_stack((a,b)) == np.vstack((a,b))
print '\n'
```

## Splitting

```python
a = np.arange(9).reshape(3, 3)
print 'a = \n', a, '\n'
```

### Horizontal / Vertical splitting

```python
# Horizontal splitting
print np.hsplit(a, 3)
print np.split(a, 3, axis=1)

# Vertical splitting
print np.vsplit(a,3)
print np.split(a, 3, axis=0)
```

### Depth-wise splitting

```python
c = np.arange(27).reshape(3,3,3)
print c
print np.dsplit(c, 3)
```

## Converting arrays

We can convert a NumPy array to a Python list with the `tolist()` function.

```python
b = np.array([1.+1.j, 3.+2.j])
print b
print b.tolist()
```

The `astype()` function transforms the array to an array of the specified data type.

```python
print b.astype('int')  # There will be a warning.
```

### Views & Copies

```python
import scipy.misc
import matplotlib.pyplot as plt

lena = scipy.misc.face()  # lena() is deprecated from SciPy. Use face() instead.
acopy = lena.copy()
aview = lena.view()

plt.subplot(231)
plt.imshow(lena)

plt.subplot(232)
plt.imshow(acopy)

plt.subplot(233)
plt.imshow(aview)

aview.flat = 0

plt.subplot(234)
plt.imshow(lena)

plt.subplot(235)
plt.imshow(aview)

plt.subplot(236)
plt.imshow(acopy)
plt.show()
```

### Indexing with a list of locations

```python
import scipy.misc
import matplotlib.pyplot as plt
import numpy as np

lena = scipy.misc.face()
xmax = lena.shape[0]
ymax = lena.shape[1]

def shuffle_indices(size):
    arr = np.arange(size)
    np.random.shuffle(arr)
    
    return arr

xindices = shuffle_indices(xmax)
np.testing.assert_equal(len(xindices), xmax)
yindices = shuffle_indices(ymax)
np.testing.assert_equal(len(yindices), ymax)
plt.imshow(lena[np.ix_(xindices, yindices)])
plt.show()
```

### Broadcasting NumPy arrays

```python
import scipy.io.wavfile
import matplotlib.pyplot as plt
import urllib2
import numpy as np

response = urllib2.urlopen('http://www.thesoundarchive.com/austinpowers/smashingbaby.wav')
print response.info()

WAV_FILE = 'smashingbaby.wav'
filehandle = open(WAV_FILE, 'w')
filehandle.write(response.read())
filehandle.close()
sample_rate, data = scipy.io.wavfile.read(WAV_FILE)

print "Data type", data.dtype, "Shape", data.shape
plt.subplot(2,1,1)
plt.title('Original')
plt.plot(data)

newdata = data * 0.2
newdata = newdata.astype(np.uint8)
print "Data type", newdata.dtype, "Shape", newdata.shape
scipy.io.wavfile.write("quiet.wav", sample_rate, newdata)
plt.subplot(2,1,2)
plt.title("Quiet")
plt.plot(newdata)
plt.show()
```
